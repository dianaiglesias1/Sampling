### Create a SMOTE sample

SMOTE works by selecting pair of minority class observations and then creating a synthetic point that lies on the line connecting these two.

```python
#I will use the implementations provided by the imbalanced-learn Python library

import imblearn
```

```python
from imblearn.over_sampling import SMOTE
```

```python
#For this, I will convert Gold to 1, 
#Silver to 2 and Bronze to 3

fgym['Medal'] = [1 if x == 'Gold' else 2 if x=='Silver' else 3 for x in fgym['Medal']]
fgym.head()
```

```python
fgym['Medal'].value_counts()
```

```python
fgym.describe().round(2)
```

```python
#dropping String columns and some that are causing trouble because not all the values

fgym = fgym.drop(columns =['Name', 'Sex', 'Team', 'NOC', 'Games', 'Season', 'City', 'Sport', 'Event'])
fgym.head()
```

```python
#applying SMOTE:

smote = SMOTE(sampling_strategy = 'minority')
X_sm, y_sm = smote.fit_resample(fgym, fgym['Medal'])
X_sm
```

```python
X_sm['Medal'].value_counts()
```

```python
X_sm.describe().round(2)
```

As we see, the count value is slightly bigger than in the original dataframe as we're oversampling with SMOTE. The category with less initial observations (2) now have more than before.

Continua amb el conjunt de dades de tema esportiu i genera una mostra utilitzant el mètode Reservoir sampling.

In general, Reservoir Sampling is well suited to situations where data arrives sequentially over time and/or is too large to fit entirely into memory.

```python
import random
```

```python
def generator(max):
    number = 1
    while number < max:
        number += 1
        yield number

stream = fgym['Medal']

k=200
reservoir = []
for i, element in enumerate(fgym['Medal']):
    if i+1<= k:
        reservoir.append(element)
    else:
        probability = k/(i+1)
        if random.random() < probability:
           
             reservoir[random.choice(range(0,k))] = element
```

```python
reservoir
```

```python
#A la funció generate he especificat 200 valors. Hi són tots?

len(reservoir)
```

```python
#create a new dataframe using the index created by the reservoir sampling
reservoir_df = fgym.iloc[reservoir]
reservoir_df.head()
```

```python
reservoir_df.describe().round(2)
```
